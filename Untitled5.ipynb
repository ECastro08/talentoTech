```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ruta = "/content/Iris.csv"
data_iris = pd.read_csv(ruta)
data_iris.head()
```

```python
iris_data = data_iris.drop(columns=['Id'])
promedio_por_especie = iris_data.groupby('Species').mean()
promedio_por_especie
```

# Distancia Euclidiana

```python
def calculate_distance_euclidiana(x1,y1,x2,y2):
  distancia_euclidiana = np.sqrt((x2-x1)**2+(y2-y1)**2)
  return distancia_euclidiana

iris_setosa = promedio_por_especie.loc['Iris-setosa']
iris_versicolor = promedio_por_especie.loc['Iris-versicolor']
iris_virginica = promedio_por_especie.loc['Iris-virginica']

distancia_euclidiana1 = calculate_distance_euclidiana(iris_setosa[0],iris_setosa[1],iris_versicolor[0],iris_versicolor[1])
distancia_euclidiana2 = calculate_distance_euclidiana(iris_setosa[0],iris_setosa[1],iris_virginica[0],iris_virginica[1])
distancia_euclidiana3 = calculate_distance_euclidiana(iris_virginica[0],iris_virginica[1],iris_versicolor[0],iris_versicolor[0])

print("Distancia euclidiana entre Iris-setosa e Iris-versicolor:", distancia_euclidiana1)
print("Distancia euclidiana entre Iris-setosa e Iris-virginica:", distancia_euclidiana2)
print("Distancia euclidiana entre Iris-virginica e Iris-versicolor:", distancia_euclidiana3)
```

```python
def plot_points_and_line(x1, y1, x2, y2):
    # Crear una figura y un sistema de ejes
    fig, ax = plt.subplots()

    # Graficar los puntos
    ax.scatter([x1, x2],[y1, y2], color='blue')

    # Graficar la línea entre los puntos
    ax.plot([x1, x2], [y1, y2], linestyle=':', color='red')

    # Etiquetar los puntos
    ax.text(x1, y1, f'({x1}, {y1})', fontsize=12, ha='right')
    ax.text(x2, y2, f'({x2}, {y2})', fontsize=12, ha='right')

    # Calcular la distancia euclidiana
    distance = calculate_distance_euclidiana(x1, y1, x2, y2)

    # Ajustar los límites del gráfico
    ax.set_xlim(min(x1, x2) - 1, max(x1, x2) + 1)
    ax.set_ylim(min(y1, y2) - 1, max(y1, y2) + 1)

    # Etiquetas de los ejes
    ax.set_xlabel('Eje X')
    ax.set_ylabel('Eje Y')

    # Título del gráfico
    ax.set_title('Distancia Euclidiana')

    # Mostrar el gráfico
    plt.grid(True)
    plt.show()


plot_points_and_line(x1, y1, x2, y2)

```

```python
# Graficar los 3 pares de especies
x = ["setosa-versicolor", "setosa-virginica", "virginica-versicolor"]
y = [distancia_euclidiana1, distancia_euclidiana2, distancia_euclidiana3]

plt.bar(x, y)
plt.xlabel("Pares de especies")
plt.ylabel("Distancia euclidiana")
plt.title("Distancia euclidiana entre pares de especies")
plt.show()
```

# Distancia Manhatan

```python
def distancia_manhatan(x1, y1, x2, y2):
  distancia = abs(x2 - x1) + abs(y2 - y1)
  return distancia

iris_setosa = promedio_por_especie.loc['Iris-setosa']
iris_versicolor = promedio_por_especie.loc['Iris-versicolor']
iris_virginica = promedio_por_especie.loc['Iris-virginica']

distancia_manh1 = distancia_manhatan(iris_setosa[0],iris_setosa[1],iris_versicolor[0],iris_versicolor[1])
distancia_manh2 = distancia_manhatan(iris_setosa[0],iris_setosa[1],iris_virginica[0],iris_virginica[1])
distancia_manh3 = distancia_manhatan(iris_virginica[0],iris_virginica[1],iris_versicolor[0],iris_versicolor[0])

print("Distancia manh entre Iris-setosa e Iris-versicolor:",distancia_manh1)
print("Distancia manh entre Iris-setosa e Iris-virginica:", distancia_manh2)
print("Distancia manh entre Iris-virginica e Iris-versicolor:", distancia_manh3)
```

```python
x = ["setosa-versicolor", "setosa-virginica", "virginica-versicolor"]
y = [distancia_manh1, distancia_manh2, distancia_manh3]

plt.plot(x, y)
plt.xlabel("Pares de especies")
plt.ylabel("Distancia Manhattan")
plt.title("Distancia Manhattan entre pares de especies")
plt.show()
```
