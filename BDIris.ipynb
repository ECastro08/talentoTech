```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ruta = "/content/Iris.csv"
data_iris = pd.read_csv(ruta)
data_iris.head()
```

```python
iris_data = data_iris.drop(columns=['Id'])
```

```python
promedio_por_especie = iris_data.groupby('Species').mean()
promedio_por_especie
```

```python
x1 = data_iris['SepalLengthCm']
y1 = data_iris['SepalWidthCm']
x2 = data_iris['PetalLengthCm']
y2 = data_iris['PetalWidthCm']
```

```python

distancia_euclidiana = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Imprimir la distancia euclidiana
print("Distancia Euclidiana entre (x1, y1) y (x2, y2):\n", distancia_euclidiana)
```

```python

sns.scatterplot(x=distancia_euclidiana, y=data_iris['Species'])

plt.title('Scatter Plot of Distances by Species')
plt.xlabel('Euclidean Distance')
plt.ylabel('Species')
# Show the plot
plt.show()
```

```python
distancia_manhattan = np.abs(x2 - x1) + np.abs(y2 - y1)

# Imprimir la distancia de Manhattan
print("Distancia de Manhattan entre (x1, y1) y (x2, y2):\n", distancia_manhattan)

sns.scatterplot(x=distancia_manhattan, y=data_iris['Species'])

plt.title('Scatter Plot of Distances by Species')
plt.xlabel('Manhattan Distance')
plt.ylabel('Species')

# Show the plot
plt.show()
```
