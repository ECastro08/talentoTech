```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# dataset

value_row1_list = ['Arauca','Cauca','Antioquia','Valle','Choco','Sucre','Narino','Meta','Casanare','Guajira'] # [num]
value_row2_list = ['arroz','cacao','cafe','cana','chontaduro','arroz','papa','yuca','trigo','name' ] # [num]
value_row3_list = [230,450,675,234,1222,780,90,645,145,53]
value_row4_list = [120,230,250,450,500,320,150,800,750,950]

data_dict = {"Department":value_row1_list,
             "Cultivo":value_row2_list,
             "Ha_Cultivada":value_row3_list,
             "Precio_Ha":value_row4_list}
print(data_dict)

# Turn a dictionary into a dataframe
df = pd.DataFrame(data_dict)

print(df)
# Call a variable
df
```

```python
def categorizar(valor):
    if valor <=100:
        return 'Baja'
    elif valor <= 400:
        return 'Normal'
    else:
        return 'Alta'

df["Categorizacion"] = df["Ha_Cultivada"].apply(categorizar)
```

```python
correlacion = df[['Ha_Cultivada', 'Precio_Ha']].corr()
sns.heatmap(correlacion, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de calor de correlaci√≥n entre Ha_Cultivada y Precio_Ha')
plt.show()
```
